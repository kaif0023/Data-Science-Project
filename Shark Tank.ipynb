```python

## Find The Equity Per Shark
## Find The Highest Amount Invested Amounted Per Shark
## Find The Highest Equity A Sharkm Gets
## Find The Most Attracted Ideas Accepted By Shark
## Find The Difference Betmeen ask Amount And amount recieved
## How Many  percent Entrepreniurs Got More Less and Equal Amount
## Find The Differnece between Ask Equity and Deal Equity
# Find How MAny Percent OF Shark Got more less and same amount of equity in succesful deals
```

```python
l=[]
l1=[]
for i in shark.columns[18:25]:
    x=shark[shark[i]==1]
    s=x["amount_per_shark"].sum()
    l.append(s)
    i=i.replace("_deal","")
    print(i,"invested",s,"Amount.")
    l1.append(i)
amount=(sorted(l,reverse=True))
amount
```

```python
df=pd.DataFrame({"shark":l1,"Total Amount Invested":l})
df
```

```python
sns.catplot(x="Total Amount Invested",kind="bar",data=df)
```

```python
l=[]
l1=[]
for i in shark.columns[18:25]:
    x=shark[shark[i]==1]
    s=x["equity_per_shark"].sum()
    l.append(s)
    i=i.replace("_deal","")
    print(i,"gets Total",s,"Equity.")
    l1.append(i)
amount=(sorted(l,reverse=True))
amount
```

```python
df=pd.DataFrame({"shark":l1,"Total Equity":l})
df
```

```python
sns.countplot(y="Total Equity",hue="shark",data=df)
```
