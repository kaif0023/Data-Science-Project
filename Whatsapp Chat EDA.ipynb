```python
import pandas as pd
import seaborn as sns
import re
import matplotlib.pyplot as plt
```

```python
f=open("Whatsapp2.txt",encoding="utf-8")
```

```python
data=f.read()
```

```python
data
```

```python
type(data)
```

```python
print(data)
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?\s-\s'
```

```python
messages=re.split(pattern,data)
messages
```

```python
type(messages)
```

```python
message=messages[1:]
```

```python
dates=re.findall(pattern,data)
```

```python
dates
```

```python
df=pd.DataFrame({"Dates":dates,"Message":message})
```

```python
df.head()
```

```python
df["Dates"]=df["Dates"].replace(" - ","")
```

```python
df.head()
```

## Checking Information On Both The Columns

```python
df.info()
```

```python
df["Dates"]=pd.to_datetime(df["Dates"],format="%m/%d/%y, %H:%M - ")
df.dtypes
```

```python
df["year"]=df["Dates"].dt.year
df["month"]=df["Dates"].dt.month_name()
df["day"]=df["Dates"].dt.day
df["hour"]=df["Dates"].dt.hour
df["hour"]=df["Dates"].dt.month
df["day_name"]=df["Dates"].dt.day_name()
df["Date"]=df["Dates"].dt.date
df
```

```python
del df["Dates"]
```

```python
df.head()
```

```python
users = []
messages = []
for i in df['Message']:
    print(i)
    if 'changed the subject' not in i:
        li = i.split(':',maxsplit=1)
    else:
        li = []
        li.append(i)
    #print(li) # comment krna
    if len(li)>=2:
        users.append(li[0])
        messages.append(li[1])
    else:
        users.append('group notification')
        messages.append(li[0])
df['user'] = users
df['messages'] = messages
df   

        
            
        
```

```python
df1=df[df["user"]!="group notification"]
df1.user.unique()
```

```python
words=[]
for i in df1["messages"]:
    li=i.split()
    words.extend(li)
words
```

```python
from urlextract import URLExtract
```

```python
extractor=URLExtract()
word=[]
for i in words:
    if len(extractor.find_urls(i))==0:
        word.append(i)
word
```

```python
len(word)
```

## Removing Media Omitted

```python
newword=[]
for i in word:
    if (("<Media"!=i) and ("omitted>"!=i)):
        newword.append(i)
newword
```
